## 2. BeautifulSoup 모듈


*(출처 : https://morioh.com/p/93eb90f9e62a)*  

웹 제작에 쓰이는 프로그래밍 언어인 "HTML"에서 데이터를 추출할 수 있도록 도와주는 모듈이다.  
HTML 과 XML 문서를 파싱하고 읽고 변형하는 데 사용한다.

**설치하기**  
`$ conda install BeautifulSoup4`  
`$ pip install BeautifulSoup4`

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
# 문자열에서 soup 생성하기
soup1 = BeautifulSoup('<html><head></head><body><h1>Hello</h1></body></html>')

# 로컬 파일에서 soup 생성하기
soup2 = BeautifulSoup(open('AIFFEL.html'))

# 웹 문서에서 soup 생성하기
soup3 = BeautifulSoup(urlopen('https://www.naver.com/'))
```

## 3. soup.get_text()
`soup.get_text()` 함수는 HTML 문서에서 모든 태그를 제거하고 텍스트 부분만 반환한다.

```python
soup1.get_text()
```

```python
soup2.get_text()
```

```python
soup3.get_text()
```

## 4. soup.find_all()
`soup.find_all()` 함수는 특정 태그를 찾는 데 사용한다.

```python
headers = soup1.find_all('h1')
headers
```

```python
a = soup2.find_all('a')
a
```

```python
firstLink = a[0]['href']
firstLink
```

```python
firstLink = a[0].get('href')
firstLink
```

```python
links = [link['href'] for link in soup2.find_all('a')]
links
```

```python
formats = soup3.find_all(['i', 'b', 'em', 'strong'])
formats
```
