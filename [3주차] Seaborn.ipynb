# 1. seaborn 라이브러리

seaborn 은 시각화를 위한 라이브러리로, 데이터프레임으로 다양한 통계 지표를 낼 수 있다.  
`import seaborn as sns` 로 seaborn 라이브러리를 임포트한다.

```python
#라이브러리 임포트
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np

#seaborn에서 제공하는 titanic 데이터 셋을 사용
titanic = sns.load_dataset('titanic')
titanic
```

# 2. seaborn 데이터셋
`sns.load_dataset()` 함수를 사용하면 seaborn 에서 제공하는 데이터셋을 사용할 수 있습니다.

```python
#seaborn에서 제공하는 titanic 데이터 셋을 사용
titanic = sns.load_dataset('titanic')
titanic
```

```python
#seaborn에서 제공하는 flights 데이터 셋을 사용
flights = sns.load_dataset('flights')
flights
```

```python
#seaborn에서 제공하는 flights 데이터 셋을 사용
tips = sns.load_dataset('tips')
tips
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

# 3. barplot()

`sns.barplot()` 함수를 사용합니다.  

| 옵션 | 입력값 |
|:-----|:-----|
| x | x축으로 나타내고 싶은 데이터셋의 컬럼명 |
| y | y축으로 나타내고 싶은 데이터셋의 컬럼명 |
| data | 데이터셋 |
| hue | x축의 값을 기준으로 세분화 하고싶은 컬럼명 |
| order | x축의 값을 기준으로 정렬 |

```python
fig = plt.figure(figsize=(10,5))
fig = sns.barplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
fig = sns.barplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
fig = sns.barplot(data=tips, x='day', y='total_bill', order=['Sun', 'Sat', 'Fri', 'Thur'])
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
g = sns.catplot(x="sex", y="total_bill",
                hue="smoker", col="time",
                data=tips, kind="bar",
                height=4, aspect=.7);
```

```python
fig = plt.figure(figsize=(10,5))
fig = sns.catplot(data=tips, x='day', y='total_bill', hue='sex', col='time', kind='bar')
plt.show()
```

# 4. boxplot()

`sns.baxplot()` 함수를 사용합니다.  

| 옵션 | 입력값 |
|:-----|:-----|
| x | x축으로 나타내고 싶은 데이터셋의 컬럼명 |
| y | y축으로 나타내고 싶은 데이터셋의 컬럼명 |
| data | 데이터셋 |
| hue | x축의 값을 기준으로 세분화 하고싶은 컬럼명 |
| order | x축의 값을 기준으로 정렬 |

```python
fig = plt.figure(figsize=(10,5))
fig = sns.boxplot(x=tips["total_bill"])
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
fig = sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
fig = sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker")
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
fig = sns.boxplot(x="day", y="total_bill", data=tips, order=['Sun', 'Sat', 'Fri', 'Thur'])
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
fig = sns.boxplot(x="day", y="total_bill", data=tips)
fig = sns.swarmplot(x="day", y="total_bill", data=tips, color='.25')
plt.show()
```

# 5. lineplot()

`sns.lineplot()` 함수를 사용합니다.  

| 옵션 | 입력값 |
|:-----|:-----|
| x | x축으로 나타내고 싶은 데이터셋의 컬럼명 |
| y | y축으로 나타내고 싶은 데이터셋의 컬럼명 |
| data | 데이터셋 |
| hue | x축의 값을 기준으로 세분화 하고싶은 컬럼명 |

```python
fig = plt.figure(figsize=(10,5))
fig = sns.lineplot(data=flights, x="year", y="passengers")
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
fig = sns.lineplot(data=flights, x="year", y="passengers", hue="month")
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
fig = sns.relplot(data=fmri, x="timepoint", y="signal", col="region", hue="event", kind="line")
plt.show()
```

| 옵션 | 입력값 |
